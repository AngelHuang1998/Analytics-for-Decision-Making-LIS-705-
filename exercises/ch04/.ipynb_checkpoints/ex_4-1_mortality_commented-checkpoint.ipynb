# Chapter 4: The Seaborn essentials for data visualization

```python
# Importing the pandas library for data manipulation
import pandas as pd
```

## Get the data

```python
# Loading pre-processed mortality data from a pickle file
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

## The basic Seaborn parameters

```python
# Importing seaborn for data visualization
import seaborn as sns
```

```python
# Creating a visualization using seaborn
# seaborn.relplot() 是 "關係圖（relational plot）"，它用來視覺化數據之間的關係，支援scatter plot和line plot。
# kind='line' => 圖的類型為折線圖（line plot）
# hue='AgeGroup'	=> 根據 AgeGroup（年齡組別）區分顏色
sns.relplot(data=mortality_data, kind='line',
            x='Year', y='DeathRate', hue='AgeGroup')
```

```python
#sns.lineplot() 直接繪製折線圖
sns.lineplot(data=mortality_data, 
             x='Year', y='DeathRate', hue='AgeGroup')
```

Change the aspect parameter for the plot in cell 6 to 2.

```python
# palette='bright'	設定顏色風格，讓 hue='AgeGroup' 的不同類別有不同的顏色
# aspect=2	設定圖表的寬高比為 2:1，讓圖表更寬
sns.relplot(data=mortality_data, kind='line', 
            x='Year', y='DeathRate', hue='AgeGroup', palette='bright',
            height=4, aspect=2)
```

```python
#col='AgeGroup'	依據 AgeGroup 拆分成多個子圖
#col_wrap=2	每列最多放 2 個子圖，自動換行
#legend=False	隱藏圖例（不顯示 hue='AgeGroup' 的顏色標籤）
sns.relplot(data=mortality_data, 
            kind='line', x='Year', y='DeathRate', hue='AgeGroup', 
            height=4, aspect=1.25, 
            col='AgeGroup', col_wrap=2, legend=False)
```

## Enhancing plots

```python
#ax.set()=設定圖表的標題 (title) 和 Y 軸標籤 (ylabel)
ax = sns.lineplot(data=mortality_data, 
                  x='Year', y='DeathRate', hue='AgeGroup')
ax.set(title='Deaths by Age Group', ylabel='Deaths per 100,000')
```

```python
#g.axes	取得 FacetGrid 內所有的子圖 axes(這邊只有一個子圖)
#g.axes.flat將多個子圖攤平成一個列表，方便 for 迴圈處理
g = sns.relplot(data=mortality_data, kind='line',
    x='Year', y='DeathRate', hue='AgeGroup', aspect=1.5)
for ax in g.axes.flat:
    ax.set(title='Deaths by Age Group', ylabel='Deaths per 100,000')
```

Change cell 10 so every fifth year is displayed on the x-axis for the plot and so the labels on the x-axis are displayed at their default rotation.

```python
# xticks 明確指定 X 軸的刻度位置，不影響顯示範圍，只影響刻度 => [1910, 1915, 1920, 1925, 1930]
# xlim=(1910, 1930) 用來 設定 X 軸的顯示範圍，讓圖表只顯示 1910-1930 之間的數據(超出範圍的數據不會顯示)

g = sns.relplot(data=mortality_data, kind='line',
    x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25)
for ax in g.axes.flat:
    ax.set(title='Deaths by Age Group (1910-1930)', ylabel='Deaths per 100,000', 
        xticks=[x for x in range(1910, 1931, 5)], 
        xlim=(1910,1930), ylim=(0,1750)),
    # ax.tick_params('x', labelrotation=30)
```

```python
ax = sns.lineplot(data=mortality_data,
     x='Year', y='DeathRate', hue='AgeGroup')
ax.set(title='Deaths by Age Group (1910-1930)', 
      ylabel='Deaths per 100,000', 
      xticks=[x for x in range(1910,1931,2)], 
      xlim=(1910,1930), ylim=(0,1750))

#旋轉 X 軸的標籤（ticks labels），讓標籤更容易閱讀
ax.tick_params('x', labelrotation=30)
```

```python
# sns.set_style => 設定 seaborn 圖表的背景樣式（style），'whitegrid' 代表 白色背景 + 網格線 
sns.set_style('whitegrid')
p = sns.relplot(data=mortality_data.query('Year >= 1910 and Year <= 1930'),
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25, )
for ax in p.axes.flat:
    ax.set(title='Deaths by Age Group (1910-1930)', ylabel='Deaths per 100,000', 
        xticks=[x for x in range(1910, 1931, 5)])
```

```python
sns.set_style('darkgrid')
sns.relplot(data=mortality_data.query('Year >= 1910 and Year <= 1930'),
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25)
```

```python
#背景為 白色，並顯示軸刻度線（ticks）
sns.set_style('ticks')
```

```python
#  g.fig.suptitle:
#  vg.fig → 取得 FacetGrid（sns.relplot() 產生的圖）
#  .suptitle() → 為整個 FacetGrid 圖表群加上標題。
#  y=1 → 控制標題的位置，y=1 代表標題放在圖的最上方
g = sns.relplot(data=mortality_data.query('Year >= 1910 and Year <= 1930'),
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', legend=False,
    col='AgeGroup', col_wrap=2, height=4, aspect = 1.25)
g.fig.suptitle('Deaths by Age Group (1910-1930)', y=1)
g.set(ylabel='Deaths per 100,000', 
      xticks=[x for x in range(1910,1931,3)], 
      ylim=(0,1750))
```

## Relational plots

```python
# # palette='colorblind'	設定顏色風格
sns.relplot(data=mortality_data, kind='line', x='Year', y='DeathRate', 
            hue='AgeGroup', palette = 'colorblind')
```

Change cell 17 so the confidence interval for the plot shows the standard deviation.

```python
#  errorbar = ('sd')   => 用來顯示標準差（Standard Deviation, SD）作為誤差範圍，讓我們可以看到數據的變異程度。
sns.relplot(data=mortality_data, kind='line', x='Year', y='DeathRate', errorbar = ('sd'))
```

```python
# scatter plot
sns.relplot(data=mortality_data, kind='scatter', x='Year', y='DeathRate', 
            hue='AgeGroup')
```

Change cell 19 so it uses a specific method to create the scatter plot.

```python
# size='DeathRate' =>讓 scatterplot 依照 DeathRate 調整點的大小
# sizes=(10,100)  => 指定點的大小範圍。最小的 DeathRate → 對應 大小 10 / 最大的 DeathRate → 對應 大小 100
g = sns.scatterplot(data=mortality_data.query('Year >= 1910 and Year <= 1930'), 
    x='Year', y='DeathRate', hue='AgeGroup', 
    size='DeathRate', sizes=(10,100))
# xticks 明確指定 X 軸的刻度位置，不影響顯示範圍，只影響刻度
g.set(xticks=[x for x in range(1910,1931,2)])
```

## Categorical plots

Change cell 20 so it uses a specific method to create the bar plot, and change the confidence interval to 75.

```python
# X 軸：只顯示 1950 和 2000
# errorbar=('ci', 75) → 使用 75% 信賴區間（Confidence Interval, CI）作為誤差線 
sns.barplot(data=mortality_data.query('Year in (1950,2000)'), 
            x='Year', y='DeathRate', errorbar = ('ci',75))
```

```python
#catplot 是用來繪製類別數據（categorical data） 的視覺化圖表，例如 長條圖（bar chart）、點圖（strip plot）、箱型圖（box plot）
#X 軸：只顯示 1950 和 2000
#orient='v'：長條圖方向為 垂直（vertical）
sns.catplot(data=mortality_data.query('Year in (1950,2000)'), 
            kind='bar', y='DeathRate', x='Year', hue='AgeGroup', orient='v')
```

```python
#kind='box' 用來 繪製箱型圖（box plot），用於顯示數據的 分佈（分位數、中位數、異常值）。
sns.catplot(data=mortality_data.query('Year >= 1915 and Year <= 1920'), 
            kind='box', x='Year', y='DeathRate')
```

```python
#畫水平的箱型圖（box plot）
sns.catplot(data=mortality_data,
            kind='box', x='DeathRate', y='AgeGroup', orient='h')
```

## Distribution plots

```python
#seaborn.displot() 用來 繪製數據的分佈圖（distribution plot），適合用來觀察數據的形態，例如：

#直方圖（Histogram, kind='hist'）
#密度曲線（Kernel Density Estimate, kind='kde'）
#累積分佈（Empirical Cumulative Distribution, kind='ecdf'）
```

```python
##直方圖（Histogram, kind='hist'）
sns.displot(data=mortality_data, kind='hist', x='DeathRate')
```

Change cell 25 so the histogram uses twice as many bins as it currently does

```python
# bins 用來設定數據區間的數量，決定直方圖的細緻程度。
sns.displot(data=mortality_data, kind='hist', x='DeathRate', bins=16)
```

```python
#密度曲線（Kernel Density Estimate, kind='kde'）
sns.displot(data=mortality_data, kind='kde', x='DeathRate')
```

```python
#累積分佈（Empirical Cumulative Distribution, kind='ecdf'）
sns.displot(data=mortality_data, kind='ecdf', x='DeathRate', hue='AgeGroup')
```

```python
# 繪製直方圖（Histogram），並且加入密度估計（KDE, Kernel Density Estimation）
sns.displot(data=mortality_data, kind='hist', 
            x='DeathRate', kde=True, bins=8)
```

```python
# 使用 sns.displot() 繪製核密度估計（KDE）圖，並根據 AgeGroup拆分為多個子圖
# fill=True → 填充 KDE 曲線底部
sns.displot(data=mortality_data, kind='kde', x='DeathRate', hue='AgeGroup', 
            fill=True, col='AgeGroup', col_wrap=2, height=3, legend=False)
```

## Other techniques for enhancing a plot

```python
g = sns.relplot(
    data=mortality_data, 
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25)
for ax in g.axes.flat:
    ax.set_title('Deaths by Age Group (2000-2016)')
    ax.set_xlabel('')
    ax.set_xticks([x for x in range(2000,2017,2)])
    ax.set_xticklabels(['Year ' + str(x) for x in range(0,17,2)])
    ax.tick_params('x', labelrotation=30)
    ax.set_ylabel('Deaths per 100,000')
    ax.set_xlim(2000,2016)
    ax.set_ylim(0,80)
```

```python
ax = sns.lineplot(data=mortality_data, 
                  x='Year', y='DeathRate', hue='AgeGroup')
ax.set_title('Deaths by Age Group')
ax.set_ylabel('Deaths per 100,000')

#ax.annotate用來作註解
#text='Spanish Flu Pandemic' → 註解的內容
#xy=(1918, 1650) → 箭頭指向 (X=1918, Y=1650)（1918 年的死亡率數值）
#xytext=(1925, 1900) → 文字標註位置 (X=1925, Y=1900)（放在 1925 年較高的位置）
#arrowprops=... → 箭頭的樣式 (facecolor='red' → 箭頭顏色為紅色 / width=3 → 箭身寬度 / headwidth=12 → 箭頭寬度 / headlength=6 → 箭頭長度)
ax.annotate(text='Spanish Flu Pandemic', 
    xy=(1918, 1650), xytext=(1925, 1900), 
    arrowprops=dict(facecolor='red', width=3, headwidth=12, headlength=6))
```

```python
# 自行設定sns.set_palette的顏色
colors = ['black','red','orange','blue']
sns.set_palette(sns.color_palette(colors))
sns.relplot(data=mortality_data, kind='line',
            x='Year', y='DeathRate', hue='AgeGroup')
```

```python
g = sns.relplot(
    data=mortality_data.query('Year >= 1910 and Year <= 1930'),
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', legend=False,
    col='AgeGroup', col_wrap=2, height=3, aspect=1.2)
g.fig.suptitle('Deaths by Age Group (1910-1930)', y=1.025)
for ax in g.axes.flat:
    ax.set_ylabel('Deaths per 100,000')
    ax.set_xticks([x for x in range(1910, 1931, 2)])
    ax.tick_params('x', labelrotation=45)   
```

```python
# Creating a visualization using seaborn
g = sns.relplot(
    data=mortality_data.query('Year >= 1910 and Year <= 1930'), 
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', legend=False, 
    col='AgeGroup', col_wrap=2, height=3, aspect=1.2)
g.fig.suptitle('Deaths by Age Group (1910-1930)', y=1.025)

#從 mortality_data 中提取 AgeGroup 的唯一值 
#(drop_duplicates()：去除重複值
#tolist()：轉換為 Python 的 list，最終 ageGroups 會是一個列表。例如:['01-04', '05-09', '10-14', '15-19']
ageGroups = mortality_data['AgeGroup'].drop_duplicates().tolist()

#enumerate() 是 Python 內建函式，用來在 for 迴圈中同時取得索引值 (index) 和對應的元素 (value)。
for index, ax in enumerate(g.axes.flat):
    ax.set_title(ageGroups[index])
    ax.set_ylabel('Deaths per 100,000')
    ax.set_xticks([x for x in range(1910, 1931, 2)])
    ax.tick_params('x', labelrotation=45)
```

```python
# import a Matplotlib module and set the figure size
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,7.5)) 

# create the line plot
sns.lineplot(data=mortality_data, x='Year', y='DeathRate', hue='AgeGroup')

# use the ax and fig objects to enhance and save the plot
ax.set(title='Deaths by Age Group (1910-1930)', 
    ylabel='Deaths per 100,000', 
    xticks=[x for x in range(1910, 1931, 2)], 
    xlim=(1910,1930), ylim=(0,1750)),
ax.tick_params('x', labelrotation=45)
```

Create a plot that contains four bar subplots that display the death rates by age group for the years 1900, 1925, 1950, 1975, and 2000. Display two subplots in each row, and set the height of the plot to an appropriate size. Add a title to the plot, and set the label for the y-axis to “Deaths per 100,000”. Note the position of the title, and then fix it so it’s displayed above the titles for the subplots. Save the plot to a file named barCharts.png in the same folder as the Notebook.

```python
g = sns.catplot(
    data=mortality_data.query('Year in (1900, 1925, 1950, 1975, 2000)'), 
    kind='bar', x='Year', y='DeathRate', hue='Year', palette = 'colorblind', legend = False,
    col='AgeGroup', col_wrap=2, height=4, aspect=1.2)
g.fig.suptitle('Deaths by Age Group (1900, 1925, 1950, 1975, 2000)', y=1.025)
ageGroups = mortality_data['AgeGroup'].drop_duplicates().tolist()

#enumerate() 是 Python 內建函式，用來在 for 迴圈中同時取得索引值 (index) 和對應的元素 (value)。
for index, ax in enumerate(g.axes.flat):
    ax.set_title(ageGroups[index])
    ax.set_ylabel('Deaths per 100,000')

g.savefig('barCharts.png') #預設儲存到目前的工作目錄
```
