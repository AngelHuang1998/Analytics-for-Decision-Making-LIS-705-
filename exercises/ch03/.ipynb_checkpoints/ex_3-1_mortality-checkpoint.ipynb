# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head(5)
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
# 3. create a line plot for just the data in the 15-19 age group. 
# Include an appropriate title on the plot and remove the legend.
mortality_wide["15-19 Years"].plot.line(title='Mortality Rate for Age Group 15-19 Years',legend=False) 
```

```python
# create an area plot for all age groups
mortality_wide.plot.area()
```

```python
# 4. reverse the order of the items in the legend
reverse = mortality_wide[["15-19 Years","10-14 Years","05-09 Years","01-04 Years"]]
reverse.plot.area()
```

```python
# 5. Use the "wide data" to create a bar plot for all age groups that shows the mortality rates for just the year 1900
# bar plot : .plot.bar()
mortality_wide.query('Year == 1900').plot.bar()
```

```python
# 6. bar plot for all age groups that shows the mortality rates for the year 2000
bar = mortality_wide.query('Year == 2000').plot.bar(title = 'Mortality rates for the year 2000', xlabel = '')
```

```python
# 7. Use the long data to create a bar plot like the one in step 6
# use pivot: long data -> wide data
mortality_data.pivot(index="Year",columns="AgeGroup",values="DeathRate").query('Year == 2000').plot.bar()
```

```python
# 8. Use the long data to create a histogram that shows the frequency of the death rates in the default number of bins
# historgram: .plot.hist()
mortality_data.plot.hist(y='DeathRate')
```

```python
# 8. bins: 10(default) -> 15
mortality_data.plot.hist(y='DeathRate',bins = 15)
```

```python
# 9. Use the long data to create a density plot that shows the distribution of the death rates in the year 1900
# density plot: .plot.density()
dens_plot = mortality_data.query('Year == 1900').plot.density(y='DeathRate',title = 'Density Plot of Death Rates for the Year 1900', grid=True)
```

```python
# 10. Use the long data to create a pie plot that shows the sum of the death rates for the years 1900, 1925, 1950, 1975, and 2000.
# pie plot: .plot.pie()
mortality_data.groupby('Year')['DeathRate'].sum().loc[[1900,1925,1950,2000]].plot.pie()
```

```python
# 11. Create a plot with four subplots in two rows and two columns. 
# subplots
mortality_wide.loc[[1900, 1925, 1950, 1975, 2000]].plot.barh(
    title=['Child Mortality: 01-04','Child Mortality: 05-09',
           'Child Mortality: 10-14','Child Mortality: 15-19'],
    ylabel='Deaths per 100,000', sharey=True,
    grid=True, rot=45, legend=False,
    subplots=True, layout=(2,2), figsize=(10,10))

#sharey=True :Ensures that all subplots share the same Y-axis scale, making comparisons easier
#rot=45	:Rotates the X-axis labels by 45 degrees for better readability.
#legend=False : Hides the legend, useful when each subplot represents a single category
#subplots=True :Creates separate subplots for each column in the DataFrame, instead of plotting all data on one chart.
#figsize=(10,10) :Sets the overall figure size in inches (width, height).
```
