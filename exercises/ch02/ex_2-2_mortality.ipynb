## Get the long and wide DataFrames

```python
#3. Run the cells that import the Pandas module and get the data from the file named mortality_data.pkl.
mortality_data = pd.read_pickle('mortality_data.pkl')
mortality_data.head()
```

```python
# 4. Run the cell that gets the data from the file named mortality_wide.pkl.
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Work With the long DataFrame

```python
# 5. Display the first five rows of the long DataFrame
mortality_data.head()
```

```python
# 6. Change the name of the “DeathRate” column to “Deaths/100K
# .rename()
mortality_data.rename(columns={"Death Rate": "Deaths/100K"}, inplace=True)
mortality_data.head()
```

```python
# add a column named MeanCentered to a DataFrame
# MeanCentered save the value of "Deaths/100K - Deaths/100K mean"
#column names with special characters like / cannot be accessed using dot notation (.). Instead, you should use bracket notation ([]).
mortality_data['MeanCentered'] = mortality_data['Deaths/100K'] - mortality_data['Deaths/100K'].mean()
```

```python
# 7. display the first five rows of the Year and MeanCentered columns
mortality_data[['Year','MeanCentered']].head()
```

```python
# 8. display the last six rows of data from 1915 through 1920
mortality_data.query("1915 <= Year <= 1920").tail(6)
```

```python
# 9. display the Year and Deaths/100K columns for the age group 01-04 Years.
mortality_data.query("`Age Group` == '1-4 Years'")[['Year','Deaths/100K']]
```

```python
# 10. Sort the DataFrame by the Deaths/100K column in descending sequence, and display the first three rows of the results.
sorted_data = mortality_data.sort_values("Deaths/100K", ascending = False)
sorted_data.head(3)
```

```python
# 10. display the last three rows of the results.
sorted_data.tail(3)
```

```python
# 11. the median of all of the values in the Deaths/100K column.
mortality_data['Deaths/100K'].median()
```

```python
# 12. Group the data by year, and calculate the sum of the Deaths/100K column.
mortality_data.groupby('Year')["Deaths/100K"].sum()
```

## Work with the wide DataFrame

```python
# 13. Display the first five rows of the DataFrame.
mortality_wide.head()
```

```python
# 14. Display the index information for the DataFrame.
mortality_wide.index
```

```python
# 15. display statistical information for the numeric columns
mortality_wide.describe()
```

```python
# 15. add the T property to see how the display changes
mortality_wide.describe().T
```

```python
#  16. Access and display just the Year and 01-04 Years columns.
mortality_wide[['Year','01-04 Years']]
```

```python
# 17. Access and display just the rows for the years from 1915 through 1920.
mortality_wide.query("1915 <= Year <= 1920")
```

```python
# 18. displays the Year and 01-04 Years columns for the years from 1915 through 1920.
mortality_wide.query("1915 <= Year <= 1920")[['Year','01-04 Years']]
```

```python
#  19. Aggregate the data for all numeric columns in each year, and display the mean, median, and sum for those columns.
mortality_wide.groupby("Year").agg(["mean", "median", "sum"])
```

```python
# 20. Add a new column to the DataFrame named TotalDeaths. The value of this column should be the sum of the values in each of the year range columns. 
mortality_wide['TotalDeaths']= mortality_wide.iloc[:, 1:].sum(axis=1)  # Operate along the row direction (horizontally), that is, operate on "each row".
mortality_wide.head()
```

```python
# 21. Create a line plot that shows the total death rates by year.
mortality_wide.plot(x="Year", y="TotalDeaths")
```
