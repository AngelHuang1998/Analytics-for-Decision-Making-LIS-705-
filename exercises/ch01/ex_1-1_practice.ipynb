## Get the data

```python
import pandas as pd  # Import pandas library for data manipulation and rename panda as pd
poll_url = 'http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'   #Define the URL
polls = pd.read_csv(poll_url)  # Read the CSV
polls
```

## Sort the data

```python
#Sort the startdate from large to small. 'inplace=True' means modifying the original DataFrame and not returning a new sorted DataFrame.
polls.sort_values('startdate', ascending=False, inplace=True)
polls.head()
```

## Use lists, slices, tuples, and dictionary objects

### A list

```python
#drop() is used for delete
polls.drop(columns=['cycle','branch','matchup','forecastdate'], inplace=True) # Delete the specified columns
polls.head(2)
```
