## Get the data

```python
import pandas as pd # Import pandas library for data manipulation and rename panda as pd
poll_url = 'http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv' #Define the URL
polls = pd.read_csv(poll_url) # Read the CSV
```

```python
polls # prints all dataframe
polls.head()  # Display the first 5 rows (the default is 5)
```

```python
# Sort the DataFrame by the 'state' column
polls.sort_values('state', inplace=True) # "inplace=True" means modify the original DataFrame directly instead of returning a new DataFrame
polls.head(2) # Display the first 2 rows
```

## Use Tab completion and tooltips

```python
# Sort the DataFrame by the 'state' column in descending order, after that, show the top 5 rows
polls.sort_values('state', ascending=False).head()  
```

## Clean the data

### Keep only the rows with "now-cast" in the type column

```python
polls = polls.query('type=="now-cast"')  # keep only the rows where 'type' is "now-cast"  
polls.head() # Display the first 5 rows (the default is 5)
```

### Rename voting columns

```python
# rename 'rawpoll_clinton' to 'Clinton_pct'
# rename 'rawpoll_trump' to 'Trump_pct'
polls.rename(columns={'rawpoll_clinton':'Clinton_pct',
                      'rawpoll_trump':'Trump_pct'},inplace=True)
polls.head() # Display the first 5 rows (the default is 5)
```

### Convert date columns to datetime objects

```python
date_cols = ["startdate","enddate"]   # Define a list of columns "startdate" and "enddate"
polls[date_cols] = polls[date_cols].apply(pd.to_datetime) # convert these two columns to datetime format 
polls.head() # Display the first 5 rows (the default is 5)
```

### Try two plots

```python
# draw a line chart
polls.plot.line(x='enddate',y=['Clinton_pct','Trump_pct'])  # x-axis = enddate, y-axis = 'Clinton_pct' and 'Trump_pct'
```

```python
%%time
(polls.query('state == "U.S."')     # keep only the rows where 'state' is "U.S."  
    .plot(x='enddate',y=['Clinton_pct','Trump_pct']))  # draw a line chart
```
