# Chapter 1: Introduction

## Get the data

```python
import pandas as pd

#定義 CSV 文件的 URL
poll_url = 'http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'

#pd.read_csv() 是 Pandas 中用於讀取 CSV 文件的函數
#將指定的 CSV 文件內容讀入一個 Pandas DataFrame（表格結構）中，並存儲在變數 polls 中。
polls = pd.read_csv(poll_url)
polls
```

## Sort the data

```python
#將startdate由大到小排序。inplace=True表示修改原 DataFrame，不返回排序後的新 DataFrame。
polls.sort_values('startdate', ascending=False, inplace=True)
polls.head(6)  #顯示 前 6 行
```

## Use lists, slices, tuples, and dictionary objects

### A list

```python
#drop()用來刪除
polls.drop(columns=['cycle','branch','matchup','forecastdate'], inplace=True) #刪除指定column
polls.head(2)
```

### A tuple

```python
#plot.line()繪製折線圖
#xlim=('2016-06', '2016-11')設置 x 軸的顯示範圍
polls.plot.line(xlim=('2016-06','2016-11'))
```

### A dictionary

```python
polls.rename(columns={'adjpoll_clinton':'Clinton',
                      'adjpoll_trump':'Trump'})
```

### Two slices

```python
#選擇的列範圍從0,10.20,...100。
#行從 state 行到 grade 行
polls.loc[0:100:10,'state':'grade']
```

## How to code a list comprehension

```python
#每次從 range(1900, 1920, 2) 中取出一個數字 x，並將它加入到列表中
xticks = [x for x in range(1900,1920,2)]
xticks
```

## How to continue statements

### With implicit continuation

```python
#先按 state 列排序，若state 值相同，再按 startdate 列排序
polls.sort_values(['state','startdate'], 
                  ascending=False,
                  inplace=True)
polls.head()
```

### With explicit continuation

```python
polls.sort_values(['state','startdate'], \     #\ 為"行續符"
                  ascending=False, \
                  inplace=True)
polls.head()
```

## How to use the Magic Commands

```python
#%time 用來測量"單行"代碼的執行時間
%time polls = pd.read_csv(poll_url)
polls
```

```python
%%time   #測量的是"整個"Cell 的執行時間
polls = polls.sort_values('startdate', ascending=False)
polls
```
